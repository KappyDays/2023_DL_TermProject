```python
import os
from tqdm.auto import tqdm
from imagededup.methods import CNN

# CNN 기반 imagededup 사용
myencoder = CNN()
```

```python
name_list = ['참취', '곰취', '꽃개오동', '개오동']
remove_threshold = 0.9

path = 'real_data/01.데이터/'
for name in name_list:
    for folder in ['train/잎', 'valid/잎']:
        image_dir = path + name + '/' + folder + '/'
        print(image_dir)
        print(name + '/' + folder + " 중복 검사 중")
        
        image_num = len(os.listdir(image_dir))
        remove_list = myencoder.find_duplicates_to_remove(image_dir=image_dir, min_similarity_threshold=remove_threshold)
        remove_num = len(remove_list)
        
        print("\n중복 제거 시작\n")
        for idx, file in enumerate(tqdm(remove_list)):
            os.rename(image_dir + file, image_dir + str(idx) + 'Remove_' + file)

        # 중복 제거된 내용 기록
        with open('datas.txt', 'a') as f:
            f.write(name + '-' + folder + '=> ' + str(image_num) + '개 중 ' + str(remove_num) + '개 제거됨\n')
    
```
